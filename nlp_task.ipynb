## Positive score

```python
def pos_sentiment_scores(sentence):
  try:
    sid_obj = SentimentIntensityAnalyzer()
    sentiment_dict = sid_obj.polarity_scores(sentence)
    return sentiment_dict['pos']*100
  except:
    return None
```

## Subjectivity and Polarity

```python
from textblob import TextBlob
```

```python
def subjectivity(text):
  try:
    analysis = TextBlob(text)
    return analysis.sentiment.subjectivity
  except:
    return None
```

```python
def polarity(text):
  try:
    analysis = TextBlob(text)
    return analysis.sentiment.polarity
  except:
    return None
```

```python
df['subjectivity_score'] = df['Text'].apply(subjectivity)
```

```python
df['polarity_score'] = df['Text'].apply(polarity)
```

## Avg Sentence Length

```python
from nltk.tokenize import sent_tokenize

```

```python
nltk.download('punkt')
```

```python
def avg_len(text):
  try:
    s = sent_tokenize(text)
    count = len(s)
    len_sum = 0
    for i in s:
      len_sum+=(len(i.split()))
    return len_sum/count
  except:
    return None
```

```python
df['average_length_of_sentence'] = df['Text'].apply(avg_len)
```

## PERCENTAGE OF COMPLEX WORDS

```python
from nltk.tokenize import word_tokenize
```

```python
def complex_percent(text):
  try:
    text.lower()
    words = word_tokenize(text)
    vowels = ['a','e','i','o','u']
    for i in range(len(words)):
      if words[i][-2:]=='es':
        words[i] = words[i].replace('es','')
      if words[i][-2:]=='ed':
        words[i] = words[i].replace('ed','')

    total = len(words)
    complex_count = 0

    for i in words:
      for j in vowels:
        if j in i:
          complex_count+=1
          break
    return (complex_count/total) *100

  except:
    return None
```

```python
complex_percent('cares stares looked gvi mkl pkl cf')
```

```python
df['Percentage_of_complex_words'] = df['Text'].apply(complex_percent)
```

## FOG Index

```python
def Fog_index(text):
  try:
    text.lower()
    words = word_tokenize(text)
    vowels = ['a','e','i','o','u']
    for i in range(len(words)):
      if words[i][-2:]=='es':
        words[i] = words[i].replace('es','')
      if words[i][-2:]=='ed':
        words[i] = words[i].replace('ed','')

    total = len(words)
    complex_count = 0

    for i in words:
      for j in vowels:
        if j in i:
          complex_count+=1
          break
    return (complex_count/total) 

  except:
    return None
```

```python
df['FOG_index'] = df['Text'].apply(Fog_index)
```

## AVG NUMBER OF WORDS PER SENTENCE


```python
def avg_word(text):
  try:
    words = word_tokenize(text)
    sent = sent_tokenize(text)
    return len(words)//len(sent)
  except:
    return None
```

```python
df['AVG NUMBER OF WORDS PER SENTENCE'] = df['Text'].apply(avg_word)
```

## COMPLEX WORD COUNT

```python
def complex(text):
  try:
    text.lower()
    words = word_tokenize(text)
    vowels = ['a','e','i','o','u']
    for i in range(len(words)):
      if words[i][-2:]=='es':
        words[i] = words[i].replace('es','')
      if words[i][-2:]=='ed':
        words[i] = words[i].replace('ed','')

    complex_count = 0

    for i in words:
      for j in vowels:
        if j in i:
          complex_count+=1
          break
    
    return complex_count

  except:
    return None
```

```python
df['COMPLEX WORD COUNT'] = df['Text'].apply(complex)
```

## WORD COUNT

```python
from nltk.corpus import stopwords
import re
import string
```

```python
nltk.download('stopwords')
```

```python
stop_words = stopwords.words('english')
```

```python
def word_count(text):
  try:
    text = text.lower()
    text = re.sub(f'[{string.punctuation}]','',text)

    words = word_tokenize(text)
    words = [word for word in words if word not in stop_words]
    return len(words)
  except:
    return None
```

```python
df['Word_count'] = df['Text'].apply(word_count)
```

```python
df.head()
```

## SYLLABLE PER WORD


```python
def s_words(text):
  try:
    words = word_tokenize(text)
    vowels = ['a','e','i','o','u']
    counts = 0
    for i in words:
      for j in i:
        if j in vowels:
          counts+=1
    return counts/len(words)

  except:
    return None
```

```python
df['SYLLABLE PER WORD'] = df['Text'].apply(s_words)
```

## Personal pronouns

```python
def person(text):
  try:
    pronouns = ['I','we','my','ours','us']
    words = word_tokenize(text)
    counts = 0
    for i in words:
      if i in pronouns:
        counts+=1
        
    return counts

  except:
    return None
```

```python
df['Personal_pronouns'] = df['Text'].apply(person)
```

## Average Word Length


```python
def w_len(text):
  try:
    new_text = text.strip()
    char_len = len(new_text)
    words = word_tokenize(text)
    word_len = len(words)
    return char_len//word_len

  except:
    return None
```

```python
df['Average Word Length'] = df['Text'].apply(w_len)
```
